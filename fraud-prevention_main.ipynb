# Data Correction 

```python
### numerical data

int_data = train_data.select_dtypes(include = int)
float_data = train_data.select_dtypes(include = float)
numerical_flags = list(int_data.keys()) + list(float_data.keys())

### categorical data 
categorical_flags = ['ad_content_price_currency','ad_content_category_name',
                     'ad_content_customerSpecific_device','ad_content_customerSpecific_autoColorM',
                     'ad_location_city', 'ad_location_region','ad_content_customerSpecific_wheelSize',
                     'ad_content_customerSpecific_wheelPcd','ad_content_customerSpecific_autoMotoModel']

### Date data 
date_flags = ['ad_content_updatedAt', 'ad_content_createdAt', 'ad_user_customerSpecific_registeredAt']

###  text data 
text_flags = ['ad_content_title', 'ad_content_body']

### target
targets = train_data['ad_customerSpecific_moderationDecision']
```

```python
# discovering missing values and drop useless features having more than 95 percent of missing values 
data = train_data

total = data.isnull().sum().sort_values(ascending = False)  
percent = (data.isnull().sum() / data.isnull().count() * 100 ).sort_values(ascending = False) 
to_drop = [total.keys()[i] for i  in range(len(total)) if percent[i] > 95 ]

#droping somme features and reduce data size 
num_flags = [ c for c in numerical_flags if not c in to_drop]
cat_flags = [ c for c in categorical_flags if not c in to_drop]
txt_flags = [ c for c in text_flags if not c in to_drop]
```

##  numerical data

```python
# for numerical data the way to go to correct missing data is sklearn Imputer 
try:
    from sklearn.impute import SimpleImputer # Scikit-Learn 0.20+
except ImportError:
    from sklearn.preprocessing import Imputer as SimpleImputer

imputer = SimpleImputer(strategy = "mean")
imputer.fit( train_data[num_flags] )

X_num = imputer.transform( train_data[num_flags] )
X_num_train = pd.DataFrame(X_num, columns = num_flags)
```

## categorical data

```python
from sklearn.preprocessing import FunctionTransformer

def categorical_missing_values(data , categorical_columns):
     # get boolean features 
    cat_data = data[ categorical_columns ].copy()
    cat_data['ad_content_customerSpecific_autoColorM'] = cat_data['ad_content_customerSpecific_autoColorM'].copy().fillna('false')
    # replace all missing value with unified one 
    return cat_data.fillna('unknown')

categoricalTransformer = FunctionTransformer(categorical_missing_values, validate = False,
                                 kw_args={"categorical_columns": cat_flags})

transformed_cat_data = categoricalTransformer.fit_transform( train_data )
```

## date data

```python
date_data = train_data[ date_flags ]
date_data.head()
```

```python
# cheking if there is any missing value 
date_data.isnull().sum().sort_values( ascending = False ) 
```

it seem's that there are a lot of missing values in ad_user_customerSpecific_registeredAt feature, the simplest way to resolve this problem is replacing the rows corresponding to the median

```python
from dateutil import parser
import numpy as np

def get_dates_median(dates):
    ''' return the median date for each column feature'''
    non_null_dates= dates[dates.isnull() == False]
    sorted_dates = list(sorted( non_null_dates ))
    return sorted_dates[len(sorted_dates)//2]

def fillna_dates( dates ):
    ''' search for data columns having missing values and replace them by the median date'''
    dates_fct = dates.copy()
    cat_contains_nan = [c for c in dates_fct.isnull().sum().keys() if  dates_fct.isnull().sum()[c] > 0 ]
    for cat in cat_contains_nan:
        date = dates_fct[ cat ]
        date_median = get_dates_median( date ) 
        dates_fct[ cat ] = dates_fct[ cat ].fillna( date_median )
    return dates_fct

def dates_missing_values(data, date_columns):
    '''convert dates into secondes and get day times differences between dates in data'''
    # extracting dates from data as Timestamp
    
    date_data = fillna_dates( data[date_columns].copy() ) 
    update_day = date_data['ad_content_updatedAt'].apply(lambda e: int(parser.parse(e).strftime("%s")) / 86400)
    created_day = date_data['ad_content_createdAt'].apply(lambda e: int(parser.parse(e).strftime("%s")) / 86400)
    regitered_day = date_data['ad_user_customerSpecific_registeredAt'].apply(lambda e: int(parser.parse(e).strftime("%s")) / 86400)
   
    # creating the dates dataframes 
    date_column_names = ['created_days_diff','updated_days_diff']
    created_days_diff = created_day - regitered_day
    updated_days_diff = update_day - regitered_day
    merged_date_data = pd.DataFrame({'created_days_diff': created_days_diff ,
                                    'updated_days_diff': updated_days_diff})
    return merged_date_data

dateTransformer = FunctionTransformer(dates_missing_values, validate = False,
                                 kw_args={"date_columns": date_flags})
dates_transformed = dateTransformer.fit_transform( train_data )
```

## preparing data for training 

```python
try:
    from sklearn.preprocessing import OrdinalEncoder # just to raise an ImportError if Scikit-Learn < 0.20
    from sklearn.preprocessing import OneHotEncoder
except ImportError:
    from future_encoders import OneHotEncoder  
    
target_encoder = OneHotEncoder()
target_encoded = target_encoder.fit_transform( np.asarray(targets).copy().reshape(-1,1) ).toarray()
y = target_encoded[:,0]
```

```python
# creating final dataset
text_data = train_data[ text_flags ].copy().fillna("unknown")

X  = pd.DataFrame()
X[ list(X_num_train.keys()) ] = X_num_train
X[list(transformed_cat_data.keys())] = transformed_cat_data
X[list(dates_transformed.keys())] = dates_transformed
X[list(text_data.keys())] = text_data
```

encode categorical features. CatBoost supports both numerical and categorical features. However, we do need to identify categorical features indices.

```python
def column_index(df, query_cols):
    """encode categorical features and identify categorical features indices"""
    cols = df.columns.values
    sidx = np.argsort(cols)
    return sidx[np.searchsorted(cols, query_cols, sorter=sidx)]
```

```python
categorical = [list(transformed_cat_data.keys()) + list( text_data )]
categorical_features_indices = column_index(X, categorical)
ct = categorical_features_indices[0]
```

CatBoost is a machine learning algorithm developed by Yandex, uses gradient boosting on decision trees.
It is available as an open source library.

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.25, random_state=42)
```

### training the model

```python
from catboost import CatBoostClassifier
model = CatBoostClassifier(
    iterations=50,
    depth=3,
    learning_rate=0.1,
)
```

```python
model.fit(X_train, y_train,cat_features = ct,eval_set=(X_valid, y_valid),plot=True)
```

it seems that the model gives a fair solution and training & testing error are pretty much in sync.
We will try to tune model parameters, features to improve the results.

```python
from catboost.utils import select_threshold
from catboost import Pool
print(select_threshold(model=model, data= Pool(X_valid, label=y_valid, cat_features = ct), FNR=0.01))
```

### tuning the model hyperparameters

CatBoost Model Tuning
iterations: is maximum number of trees that can be built when solving machine learning problems.
learning_rate: is used for reducing the gradient step.

```python
from catboost import cv, Pool

params = {}
params['loss_function'] = 'Logloss'
params['iterations'] = 20
params['custom_loss'] = 'AUC'
params['random_seed'] = 63
params['learning_rate'] = 0.5

cv_data = cv(
    params = params,
    pool = Pool(X, label=y, cat_features= ct),
    fold_count=3,
    shuffle=True,
    partition_random_seed=0,
    stratified=False,
)
```

## ploting validation and training losss 

```python
from catboost.utils import get_roc_curve
import sklearn
from sklearn import metrics

eval_pool = Pool(X_valid, y_valid, cat_features= ct)
curve = get_roc_curve(model, eval_pool)
(fpr, tpr, thresholds) = curve
roc_auc = sklearn.metrics.auc(fpr, tpr)

from catboost.utils import get_fpr_curve
from catboost.utils import get_fnr_curve

(thresholds, fpr) = get_fpr_curve(curve=curve)
(thresholds, fnr) = get_fnr_curve(curve=curve)


plt.figure(figsize=(16, 8))
lw = 2

plt.plot(thresholds, fpr, color='blue', lw=lw, label='FPR', alpha=0.5)
plt.plot(thresholds, fnr, color='green', lw=lw, label='FNR', alpha=0.5)

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.grid(True)
plt.xlabel('Threshold', fontsize = 16)
plt.ylabel('Error Rate', fontsize = 16)
plt.title('FPR-FNR curves', fontsize = 20)
plt.legend(loc="lower left", fontsize = 16)
plt.show()
```
